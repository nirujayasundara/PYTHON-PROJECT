```python
import pandas as pd
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display, clear_output

file = "Python_Basic_Project_With_Database.xlsx"

def load_sheet(name):
    df = pd.read_excel(file, sheet_name=name, header=2)
    df.columns = df.columns.str.strip()
    return df

payments = load_sheet("Payments")
enrollments = load_sheet("Enrollments")
batches = load_sheet("Batches")
courses = load_sheet("Courses")
students = load_sheet("Students")
```

```python
df = payments.merge(enrollments, on='enroll_id') \
             .merge(batches, on='batch_id') \
             .merge(courses, on='course_id')
```

```python
total_revenue = df['amount'].sum()
total_students = df['student_id'].nunique()
total_courses = df['course_name'].nunique()
```

```python
def show_kpis(data):
    revenue = data['amount'].sum()
    students = data['student_id'].nunique()
    courses = data['course_name'].nunique()

    print("📊 DASHBOARD KPI")
    print("----------------------")
    print(f"💰 Total Revenue: {revenue}")
    print(f"👨‍🎓 Students: {students}")
    print(f"📚 Courses: {courses}")
```

```python
course_dropdown = widgets.Dropdown(
    options=['All'] + list(df['course_name'].unique()),
    value='All',
    description='Course:'
)
```

```python
output = widgets.Output()

def update_dashboard(change):
    with output:
        clear_output()

        filtered = df.copy()

        if course_dropdown.value != 'All':
            filtered = filtered[filtered['course_name'] == course_dropdown.value]

        show_kpis(filtered)

        # Revenue by month
        temp = filtered.copy()
        temp['payment_date'] = pd.to_datetime(temp['payment_date'])
        temp['month'] = temp['payment_date'].dt.to_period('M')

        monthly = temp.groupby('month')['amount'].sum()

        monthly.plot(kind='line', title="Monthly Revenue")
        plt.show()

course_dropdown.observe(update_dashboard, names='value')
```

```python
display(course_dropdown, output)
update_dashboard(None)
```
